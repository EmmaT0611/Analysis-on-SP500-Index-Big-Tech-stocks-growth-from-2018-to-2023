```python
import pandas as pd
#load data
sp500_sectormean = pd.read_csv('sp500_sectormean.csv')
print(sp500_sectormean.head())
```

**Plot S&P500 Top Sectors Mean Daily Return**

```python
import matplotlib.pyplot as plt
import pandas as pd

df = sp500_sectormean.copy()

df.plot(x="date1", y=["m_industrial", "m_financials", "m_infotech", "m_healthc", "m_consumerd"])
plt.xlabel("Year")
plt.ylabel("Price")
plt.title("S&P500 Top Sectors Mean Daily Return")
plt.legend(["Industrial", "Financials", "Infotech", "Health Care", "Consumer Discretionary"])
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```

**Plot Cumulative Returns of S&P500 Sectors**

```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.dates import YearLocator

# Assuming you've already read the 'sp500_sectormean.csv' into sp500_sectormean DataFrame
data = sp500_sectormean.copy()

# Calculate cumulative returns
cumulative_returns = ((data.drop("date1", axis=1).pct_change() + 1).cumprod())

'''pct_change() function calculates the percentage change between consecutive elements in the data. 
This is done by taking the current value and dividing it by the previous value, and then subtracting 1. 
This gives the percentage change from the previous value to the current value.'''



# Plot cumulative returns
cumulative_returns["date1"] = pd.to_datetime(data["date1"])  # Convert date1 to datetime
cumulative_returns.set_index("date1", inplace=True)

fig, ax = plt.subplots(figsize=(10, 7))

cumulative_returns.plot(ax=ax)

plt.xlabel('Year', fontsize=14)
plt.ylabel('Growth of $1 investment', fontsize=14)
plt.title("Cumulative Returns of S&P500 Sectors", fontsize=16)
plt.legend(["Industrial", "Financials", "Infotech", "Health Care", "Consumer Discretionary"])
plt.xticks(rotation=45)

# Set the x-axis ticks to show the start of each year
year_locator = YearLocator()
ax.xaxis.set_major_locator(year_locator)

plt.tight_layout()
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
