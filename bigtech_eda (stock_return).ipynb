Historical returns are often associated with the past performance of a security or index, such as the S&P 500.

We study historical return data to:

1. forecast future returns
2. to estimate how a security might react in a situation
3. to estimate where future points of data may fall in terms of standard deviations

Calculating the historical return is done by subtracting the most recent price from the oldest price and divide the result by the oldest price.

Analyzing historical data can provide insight into how a security or market has reacted to a variety of different variables, from regular economic cycles to sudden, exogenous world events. To note that past results do not necessarily predict future returns. The older the historical return data, the less likely it'll be successful at forecasting returns in the future.

A historical return for a stock index such as the S&P 500 is typically measured from the open on January 1st to the market's close on December 31st to provide the annual return. Each year's annual return is compiled to show the historical return over several years. 

We will calculate the average historical return, i.e., a stock has returned an average of 10% per year for the past five years. To note that an average historical return doesn't mean that the stock price didn't correct lower in any of those years. The stock could have experienced price declines, but in the other years when the stock price rose, the gains more than offset the declines so that the average historical return was positive.

We will look at Impact of COVID-19 on Big Tech's Stock on price growth.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf


# List of tickers for big tech companies
tickers_list = ["META", "AMZN", "AAPL", "NFLX", "GOOG", "MSFT"]

# Fetch the data
bigtech_stocks = yf.download(tickers_list, '2019-1-1')['Adj Close']

plt.figure(figsize=(14, 5))
sns.set_style("ticks")

# Use bigtech_stocks DataFrame as the data source
company = bigtech_stocks
colorr = ['blue', 'olivedrab', 'black', 'firebrick', 'purple', 'orange']

for i, (ticker, clr) in enumerate(zip(tickers_list, colorr)):
    sns.lineplot(data=company, x=company.index, y=ticker, color=clr, label=ticker)

# First case of COVID-19 was reported in "December 2019" in Wuhan.
plt.axvspan(xmin=pd.to_datetime('2019-12-1'), xmax=pd.to_datetime('2022'), color='dimgray', alpha=0.25)
plt.text(x=pd.to_datetime("2019-12-1"), y=company.max().max() + 5, size="medium",
         s="*highlighted area shows the COVID-19 period")

# First COVID-19 case was reported in the US on "January 30, 2020"
plt.vlines(x=pd.to_datetime("2020-1-30"), color="red", ymin=company.min().min(), ymax=company.max().max() - 5)

s1 = "The day first Covid case reported in the US"
s2 = "and the stock prices started to increase after \nan initial decrease till the end of pandemic"

plt.annotate(text=s1 + "\n" + s2,
             xy=(pd.to_datetime("2020-1-30"), company['NFLX'].mean()),
             xytext=(pd.to_datetime("2019-1-1"), company['NFLX'].mean() + 100), size="medium",
             arrowprops=dict(facecolor="black", shrink=0.05))

plt.title("Big Tech Stock Prices From {0} to {1}".format(company.index.min().strftime('%Y-%m-%d'), company.index.max().strftime('%Y-%m-%d')),
          c="brown", alpha=0.8, size="xx-large")
sns.despine()
plt.ylabel("Stock Price (USD)")
plt.xlabel("Year")
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.legend(loc=(0.05, 1.1), ncol=5, frameon=False, prop={'size': 10})
plt.savefig('stock.png', bbox_inches='tight')
plt.show()
```

# Big Tech Cumulative Returns

```python
# Define the ticker list
import pandas as pd
tickers_list = ["META", "AMZN", "AAPL", "NFLX", "GOOG", "MSFT"]

# Fetch the data
import yfinance as yf
nbigtech = yf.download(tickers_list,'2019-1-1')['Adj Close']

# Plot all the close prices
((nbigtech.pct_change()+1).cumprod()).plot(figsize=(10, 7))

'''
.cumprod() function stands for cumulative product. It calculates the cumulative product of the modified percentage 
changes obtained from the previous step.This is used to compute the accumulated effect of all the individual
percentage changes over time.

After calculating the percentage changes, 1 is added to each of these changes. 
This is done to convert the percentage changes back into multiplier format. 
For example, a 10% increase is represented as 1.10.'''


# Show the legend
plt.legend()

# Define the label for the title of the figure
plt.title("Cumulative Returns of Big Tech's Share", fontsize=16)

# Define the labels for x-axis and y-axis
plt.ylabel('Growth of $1 investment', fontsize=14)
plt.xlabel('Year', fontsize=14)

# Plot the grid lines
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
